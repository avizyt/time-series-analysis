# _Power Transforms_

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
series = pd.read_csv("../data/airline-passengers.csv", header=0, index_col=0, parse_dates=True, squeeze=True)
```

```python
plt.figure(1)

# line plot
plt.subplot(211)
plt.plot(series)

# histogram
plt.subplot(212)
plt.hist(series)
plt.show()
```

```python
# square root transform
series = [i**2 for i in range(1,100)]

transform = series = np.sqrt(series)
plt.figure(1)

plt.subplot(211)
plt.plot(transform)

plt.subplot(212)
plt.hist(transform)
plt.show()
```

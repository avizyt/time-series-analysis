# Time Series Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temperature = pd.read_csv("../data/daily-min-temp.csv",header=0, index_col=0, parse_dates=True, squeeze=True)
```

```python
temperature.plot(color='red')
plt.show()
```

```python
temperature.plot(style='k. ')
plt.show()
```

```python
# create stacked line plot
groups = temperature.groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()
for name, group in groups:
    years[name.year] = group.values

years.plot(subplots=True, legend=False)
plt.show()
```

```python
# Histogram
temperature.hist(bins=20, color='Red')
plt.show()
```

```python
temperature.plot(kind='kde', color='Red')
plt.show()
```

```python
# box and whisker plot
years.boxplot()
plt.show()
```

```python
years = years.T
plt.matshow(years, interpolation=None, aspect='auto')
plt.show()
```

```python
plt.plotting.lag_plot(temperature)
plt.show()
```
